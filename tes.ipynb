```python
from qiskit_ibm_runtime import QiskitRuntimeService

service = QiskitRuntimeService()

# Fetch your 5 most recent workloads/jobs
print("Fetching recent jobs...")
recent_jobs = service.jobs(limit=5)

for j in recent_jobs:
    print(f"ID: {j.job_id()} | Status: {j.status()} | Backend: {j.backend().name}")

# If you see your job in this list, you can just grab it directly!
# For example, to grab the very last job you ran:
if recent_jobs:
    latest_job = recent_jobs[0]
    print(f"\nAutomatically grabbing the latest job: {latest_job.job_id()}")
    
    if latest_job.status() == 'DONE':
        result = latest_job.result()[0]
        counts = list(result.data.__dict__.values())[0].get_counts()
        print("Counts:", counts)
```

```python
from qiskit_ibm_runtime import QiskitRuntimeService

service = QiskitRuntimeService()

# Fetch the specific job from yesterday
job_yesterday = service.job('d8fdlnpvjngc73apn9f0')
print(f"Status: {job_yesterday.status()}")

if job_yesterday.status() == 'DONE':
    # job.result()[0] gets the data for the circuit run in that job
    result = job_yesterday.result()[0]
    
    # Extract the counts
    counts = list(result.data.__dict__.values())[0].get_counts()
    print("Counts for that single pixel comparison:", counts)
    
    prob_one = counts.get('1', 0) / sum(counts.values())
    print(f"Edge signal for this specific pixel: {prob_one:.3f}")
```
